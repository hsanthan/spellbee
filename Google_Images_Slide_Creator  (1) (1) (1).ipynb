```python
from pptx import Presentation, shapes
from pptx.util import Inches, Pt
from pptx.enum.text import PP_ALIGN
from google_images_download import google_images_download
import pandas as pd
import os
import numpy as np
```

```python
def get_google_images(num_of_words, words_df, column_name, all_paths, num_of_images):
    #instantiate the class
    response = google_images_download.googleimagesdownload()
    #num_of_images = 4
    paths = []
    for i in range(num_of_words):
       # i = i + start_idx
        arguments = {"keywords":words_df.loc[i][column_name],"limit":num_of_images,"print_urls":True, "format":"jpg", 'imgsize':'medium'}
        path = response.download(arguments)
        #print complete paths to the downloaded images
        print(path)
        paths.append(path)
    print(paths)
    all_paths.append(paths)
    return paths, all_paths



def create_pptx(num_of_words,sheet_name, paths, column_names, slides_per_presentation,num_of_images):
    #num_of_images = 4
    
    '''if num_of_words <= 250:
        step = num_of_words
    else:
        step = num_of_words//4'''
    step = slides_per_presentation
    
    for start_idx in range(0,num_of_words,step):
        if (start_idx + step) > num_of_words:
            step = num_of_words - start_idx
        
        ppt_name= sheet_name + '_'+ str(start_idx + 1) +'.pptx'
        prs = Presentation()
        #image counter
        cnt = 0
        

        try:
            for i in range(step):
                try:
                    
                    '''#Word and corresponding meaning at the bottom
                    slide = prs.slides.add_slide(prs.slide_layouts[8])
                    placeholder = slide.placeholders[1]
                    sp = slide.shapes[1].element
                    sp.getparent().remove(sp)
                    
                    shape = slide.shapes
                    shape.title.text = words.loc[i + start_idx]['Word']
                    shape.title.text_frame.paragraphs[0].font.size = Pt(44)
                    slide.placeholders[2].text = words.loc[i + start_idx]['Meaning']
                    slide.placeholders[2].text_frame.paragraphs[0].font.size = Pt(24)'''
                    
                    
                    #Word and its meaning at the top left
                    slide = prs.slides.add_slide(prs.slide_layouts[6])
                    width = height = Inches(8)
                    left = top =  Inches(0.0)
                    txBox = slide.shapes.add_textbox(left, top, width, height)
                    tf = txBox.text_frame

                    tf.text = words.loc[i + start_idx]['Word']
                    tf.paragraphs[0].font.size = Pt(40)
                    
                    print('before meaning')
                    if 'Meaning' in column_names:
                        p = tf.add_paragraph()
                        p.text = words.loc[i + start_idx]['Meaning']
                        p.font.size = Pt(24)
                    
                    print('after meaning')
                    if 'Origin' in column_names:
                        p = tf.add_paragraph()
                        p.text = words.loc[i + start_idx]['Origin']
                        p.font.size = Pt(18)

                  
                    img_paths = paths[i + start_idx][0].values()

                    for img in img_paths:
                        for path_index in range(num_of_images):

                            try:
                                if path_index == 0:
                                    top = Inches(1.5)
                                    left = Inches(5.25)

                                elif path_index == 1:
                                    top = Inches(2.25)
                                    left = Inches(1)

                                elif path_index == 2:
                                    top = Inches(4.25)
                                    left = Inches(5.25)


                                elif path_index == 3: 
                                    top = Inches(4.25)
                                    left = Inches(1)

                                height = Inches(3)
                                width = Inches(3)

                                #add the image to slide
                                pic = slide.shapes.add_picture(img[path_index], left, top, height=height, width=width)
                                #print('after image')
                            except Exception as e:
                                print(e)
                                continue

                    cnt += 1
                    #print(cnt)
                except Exception as e:
                    print(e)
                    continue
            #save the presentation
            prs.save(ppt_name)
            print('\n\n')
            print(f'Images saved in the ppt \"{ppt_name}\" : {cnt}')
        except Exception as e:
            print(e)
            
```
