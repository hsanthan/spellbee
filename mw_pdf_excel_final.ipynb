```python
#pat_pre = re.compile('([a-z]*- prefix)|([a-z]*- or* [a-z]*- prefix)|(-[a-z]* [a-z]* suffix)|(-[a-z]* also* -[a-z]* [a-z]* suffix)|([a-z]+- or [a-z]-  combining form)|([a-z]*- combining form)') 
#pat_pre = re.compile('([a-z]*-){1} (or)? ([a-z]+-)? (also)? ([a-z]+-)? (or)? ([a-z]+-)? (combining form){1}') # cr- or acro- also akr- or akro- combining form
#pat_pre = re.compile('([a-z]+- ){1}(or )?([a-z]+- )?([a-z]+ )?(prefix){1}') #a- prefix, a- or an- prefix
#pat_pre = re.compile('([a-z]+-){1} (prefix){1}') #a- prefix
#pat_pre = re.compile('([a-z]+- ){1}((or |also )?([a-z]+- )?([a-z]+ )?)*(prefix|combining form){1}') # most of the prefix and comb
pat_pre = re.compile('(([a-z]+- |-[a-z]+ ){1}((or |also )([a-z]+- |-[a-z]+ ))*([a-z]+ )?(prefix|combining form|suffix){1})')
text_pre = re.findall(pat_pre, all_text)
#print(text_pre[0::])
#re.split(pat_pre, all_text)
#req_token_list_with_none = re.split(pat_pre, all_text)
#print(req_token_list_with_none)
#re(pat_pre, all_text)
re_tok = []
tok_str = ''
for idx,t in enumerate(text_pre):
    #print(t[0])
    
    re_tok.append(t[0])
    #print(idx)
    #print(re.split(t[0], all_text)[0])
tok_str = '|'.join(re_tok)
#print(re_tok)
    
#print(tok_str)
#re.split(tok_str, all_text)
re_mean = re.split(tok_str, all_text)
re_meaning = []
for r in re_mean:
    r1 = r.replace('8', ']').replace('7','[').replace('9','<').replace(':','>').replace('�','')
    re_meaning.append(r1)
```
