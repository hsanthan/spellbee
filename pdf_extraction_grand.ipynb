```python
#pos = 'n,adj,adv,conj,v'.split(',')
#pos = 'n|adj|adv|conj|v'
```

```python
#token_pat = re.compile('^[a-z]+$')
token_pat = re.compile('(^[a-z]+ - \d?\.?)') #abacinate - 
#token_pat = re.compile('(^[a-z]+ - \d{0,1})')
#token_pat = re.compile('([a-z]+)(n|adj|adv|conj|v)')
```

```python
token_list = []
for t in text:
    #print(re.findall(token_pat,t))
    #print(t)
    if re.findall(token_pat,t):
        #if not re.findall(pos,t):
            token_list.append(re.findall(token_pat,t)[0])
token_list
```

```python
token_split = '|'.join(token_list)

token_split
```

```python
meaning = re.split(token_split,txt['content'])
meaning[1:]
```

```python
meaning_page_num_removed = []
for m in meaning:
    print(m)
    f = re.sub(re.compile('\\n\d{1,3}'),'',m)
    print(f)
    meaning_page_num_removed.append(f)
```
